# Zadanie Domowe: Klasyfikacja Binarna, Walidacja i Wizualizacja

Celem tego zadania jest stworzenie modelu klasyfikującego punkty na płaszczyźnie, monitorowanie procesu uczenia (overfitting) oraz ręczne stworzenie wizualizacji granicy decyzyjnej.

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Generowanie Danych

Uruchom poniższą komórkę, aby wygenerować zbiór danych syntetycznych. Nie zmieniaj tego kodu.

```python
def generate_dataset(n_samples=300):
    torch.manual_seed(42)
    # Klastry danych: Klasa 0 wokół (-2, -2), Klasa 1 wokół (2, 2)
    X0 = torch.randn(n_samples // 2, 2) - 1.5
    y0 = torch.zeros(n_samples // 2, 1)

    X1 = torch.randn(n_samples // 2, 2) + 1.5
    y1 = torch.ones(n_samples // 2, 1)

    # Łączenie i tasowanie
    X = torch.cat([X0, X1], dim=0)
    y = torch.cat([y0, y1], dim=0)
    indices = torch.randperm(n_samples)
    return X[indices], y[indices]

X_all, y_all = generate_dataset()

# Wizualizacja wygenerowanych danych
plt.scatter(X_all[:, 0], X_all[:, 1], c=y_all, cmap='bwr', alpha=0.7)
plt.title("Wygenerowane dane")
plt.show()
```

## CZĘŚĆ 1: Przygotowanie danych i Modelu

1. Podziel `X_all` i `y_all` na zbiór treningowy (80%) i walidacyjny (20%). Użyj slicing'u Pythona (np. `X[:240]`).
2. Zdefiniuj model używając `torch.nn.Sequential`.
   - Architektura: Wejście (2 cechy) -> Warstwa Liniowa -> 1 wyjście -> Sigmoid.
3. Zdefiniuj funkcję straty (`BCELoss`) i optymalizator (`SGD`, lr=0.1).

```python

X_train = X_all[:240]
y_train = y_all[:240]
X_val   = X_all[240:]
y_val   = y_all[240:]

model = nn.Sequential(
    nn.Linear(2, 1),
    nn.Sigmoid()
)

loss_fn   = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

## CZĘŚĆ 2: Pętla Treningowa (Train & Val)

1. Trenuj przez 1000 epok.
2. W każdej epoce:
   - Oblicz loss na zbiorze treningowym i zrób aktualizację wag (backward).
   - Przełącz model w tryb `.eval()` i oblicz loss na zbiorze walidacyjnym (bez liczenia gradientów!).
   - Wróć do trybu `.train()`.
3. Zbieraj wartości loss (train i val) do list, aby je potem wyrysować.

```python
train_losses = []
val_losses   = []

for epoch in range(1000):

    model.train()

    optimizer.zero_grad()

    y_pred_train = model(X_train)

    loss_train = loss_fn(y_pred_train, y_train)

    loss_train.backward()

    optimizer.step()
    # waga_nowa = waga_stara - lr * gradient

    train_losses.append(loss_train.item())
    # tensor jednoelementowy -> zwykłą liczbę

    model.eval()

    # walidacja
    with torch.no_grad():
        y_pred_val = model(X_val)
        loss_val = loss_fn(y_pred_val, y_val)

    val_losses.append(loss_val.item())

plt.plot(train_losses, label='Train')
plt.plot(val_losses,   label='Val')
plt.xlabel('Epoka')
plt.ylabel('Loss')
plt.title('Krzywe uczenia')
plt.legend()
plt.show()
```

## CZĘŚĆ 3: Wizualizacja Granicy Decyzyjnej (Decision Boundary)

Musisz napisać kod, który pokoloruje tło wykresu w zależności od tego, co przewiduje model.

**Algorytm:**
1. Stwórz siatkę punktów (grid) pokrywającą cały obszar danych (od min do max X i Y). Użyj `np.meshgrid` i `np.linspace`.
2. Spłaszcz tę siatkę do formatu, który przyjmie model (lista par [x, y]). `xx.ravel()` i `np.c_` lub `torch.cat`.
3. Zamień to na Tensor i przepuść przez swój wytrenowany model.
4. Wynik (prawdopodobieństwa) przekształć z powrotem do kształtu siatki (gridu).
5. Użyj `plt.contourf`, aby narysować obszary decyzyjne, a na wierzchu `plt.scatter` dla danych.

```python
model.eval()

# siatka punktow
x_min, x_max = X_all[:, 0].min() - 0.5, X_all[:, 0].max() + 0.5
y_min, y_max = X_all[:, 1].min() - 0.5, X_all[:, 1].max() + 0.5

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 200),  # 200 osi X
    np.linspace(y_min, y_max, 200)   # 200 osi Y
)

# splaszczenie siatki
grid = np.c_[xx.ravel(), yy.ravel()]

# tensor → model → przewidywania

grid_tensor = torch.tensor(grid, dtype=torch.float32)

with torch.no_grad():
    Z = model(grid_tensor)
    Z = Z.numpy()

# przywrcanie siatki

Z = Z.reshape(xx.shape)

# rysowanie

plt.figure(figsize=(8, 6))

plt.contourf(xx, yy, Z, levels=50, cmap='bwr', alpha=0.3)

plt.colorbar(label='Prawdopodobieństwo klasy 1')

plt.scatter(X_all[:, 0], X_all[:, 1], c=y_all, cmap='bwr',
            edgecolors='black', linewidths=0.5, alpha=0.8)

plt.title('Granica Decyzyjna')
plt.xlabel('Cecha 1')
plt.ylabel('Cecha 2')
plt.show()
```
